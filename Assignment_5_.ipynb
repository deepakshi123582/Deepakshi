```python
import numpy as np
gfg = np.matrix('4, 1, 9; 12, 3, 1; 4, 5, 6')

# Sum of all elements
total_sum = gfg.sum()
print("Sum of all elements:", total_sum)

# Sum of all elements row-wise
row_sum = gfg.sum(axis=1)
print("Row-wise sum:\n", row_sum)

# Sum of all elements column-wise
column_sum = gfg.sum(axis=0)
print("Column-wise sum:\n", column_sum)
```

```python
import numpy as np
array = np.array([10, 52, 62, 16, 16, 54, 453])

# i. Sorted array
sorted_array = np.sort(array)
print("Sorted array:", sorted_array)

# ii. Indices of sorted array
sorted_indices = np.argsort(array)
print("Indices of sorted array:", sorted_indices)

# iii. 4 smallest elements
smallest_elements = sorted_array[:5]
print("4 smallest elements:", smallest_elements)

# iv. 5 largest elements
largest_elements = sorted_array[-7:]
print("5 largest elements:", largest_elements)
```

```python
import numpy as np
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

# i. Integer elements only
integer_elements = array[array == array.astype(int)]
print("Integer elements:", integer_elements)

# ii. Float elements only
float_elements = array[array != array.astype(int)]
print("Float elements:", float_elements)
```

```python
import numpy as np

# Step a: Generate unique sales dataset
X = ord('A') + ord('B')
sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales Data:", sales)

# Step b: Compute personalized tax rate
tax_rate = ((X % 5) + 5) / 100
tax_amount = sales * tax_rate
print("Tax Amount:", tax_amount)

# Step c: Adjust sales based on discount
discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)
print("Discounted Sales:", discounted_sales)

# Step d: Expand sales data for multiple weeks
weekly_sales = np.tile(sales, (3,1))
print("Weekly Sales Data:\n", weekly_sales)

# Increase sales by 2% per week
weeks = np.array([1, 1.02, 1.04])[:, np.newaxis]
increased_sales = weekly_sales * weeks
print("Increased Sales Over Weeks:\n", increased_sales)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)

y1 = x**2                # y = x^2
y2 = np.sin(x)           # y = sin(x)
y3 = np.exp(x)           # y = e^x
y4 = np.log(np.abs(x) + 1)  # y = log(|x| + 1)

plt.figure(figsize=(10, 6))

plt.plot(x, y1, label="y = x^2", linestyle="--")
plt.plot(x, y2, label="y = sin(x)", linestyle="-.")
plt.plot(x, y3, label="y = e^x", linestyle=":")
plt.plot(x, y4, label="y = log(|x| + 1)", linestyle="-")


plt.title("Function Plots")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)
plt.legend()

plt.show()
```
