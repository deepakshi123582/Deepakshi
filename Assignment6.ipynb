```python
import numpy as np
import matplotlib.pyplot as plt
M = float(input("Enter a value for M: "))

# Generate x values from -10 to 10
x = np.linspace(-10, 10, 100)

# Compute y values
y1 = M * x**2
y2 = M * np.sin(x)

# Plot both functions
plt.plot(x, y1, label='y = M * x^2', color='Pink', linestyle='--')
plt.plot(x, y2, label='y = M * sin(x)', color='Black', linestyle='-')

# Add title, legend, and grid
plt.title("Plot of y = M*x^2 and y = M*sin(x)")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subject': ['Science ', 'History', 'Java', 'C++', 'DBMS'],
    'Score': [95, 80, 88, 98, 66]
}

df = pd.DataFrame(data)

plt.figure(figsize=(9, 5))
ax = sns.barplot(x='Subject', y='Score', data=df, palette='muted')

# Add annotations on top of each bar
for i, score in enumerate(df['Score']):
    ax.text(i, score + 1, str(score), ha='center', fontsize=12)

# Add title, axis labels, and grid
plt.title("Scores in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = 89654
np.random.seed(roll_number)

data = np.random.randn(50)
cumulative_sum = np.cumsum(data)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line plot showing cumulative sum
axes[0, 0].plot(cumulative_sum, color='blue', marker='o', linestyle='-')
axes[0, 0].set_title("Cumulative Sum Line Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)

# Scatter plot with random noise
axes[0, 1].scatter(range(50), data, color='red', alpha=0.6)
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Random Value")
axes[0, 1].grid(True)

axes[1, 0].axis("off")
axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("company_sales_data.csv")

# Display the first few rows
print(df.head())
```

```python
df.info()
df.describe()
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))

# Plot total profit vs month number
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', color='b')

plt.title("Total Profit of All Months")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.xticks(df['month_number'])
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(12, 6))

sns.lineplot(x=df['month_number'], y=df['facecream'], label="Face Cream", marker='o')
sns.lineplot(x=df['month_number'], y=df['facewash'], label="Face Wash", marker='o')
sns.lineplot(x=df['month_number'], y=df['toothpaste'], label="Toothpaste", marker='o')
sns.lineplot(x=df['month_number'], y=df['bathingsoap'], label="Bathing Soap", marker='o')
sns.lineplot(x=df['month_number'], y=df['shampoo'], label="Shampoo", marker='o')
sns.lineplot(x=df['month_number'], y=df['moisturizer'], label="Moisturizer", marker='o')

plt.title("Product Sales Data (Multiline Plot)")
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.xticks(df['month_number'])
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(12, 6))

df.set_index("month_number").plot(kind="bar", figsize=(12, 6), colormap="viridis", alpha=0.8)

plt.title("Sales Data for All Attributes")
plt.xlabel("Month Number")
plt.ylabel("Values")
plt.xticks(rotation=0)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```
