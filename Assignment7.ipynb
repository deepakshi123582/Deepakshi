```python
import numpy as np
import pandas as pd
roll_number = 102317100
np.random.seed(roll_number)
# Define number of months and categories
months = 12
categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]

# Generate random sales data (12 months × 4 categories)
sales_data = np.random.randint(1000, 5001, size=(months, len(categories)))

# Print generated sales data
print(sales_data)
df = pd.DataFrame(sales_data, columns=categories)
# Add month names as index
df.index = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
print(df)


```

```python
# Display first 5 rows
print(df.head())
print(df.describe())
#Calculate the total sales per category and total sales per month.
total_sales_per_category = df.sum()
print("Total sales per category:\n", total_sales_per_category)
# Total sales per month
df["Total Sales"] = df.sum(axis=1)
print("Data with Total Sales column:\n", df)
#Calculate the average sales growth between consecuƟve months for each category.
# Calculate month-over-month growth rate
df["Growth Rate"] = df["Total Sales"].pct_change() * 100
print("Data with Growth Rate column:\n", df)
#5. Apply a randomized discount:
#If your roll number is even, apply a 10% discount to the Electronics category.
#If odd, apply a 15% discount to the Clothing category.
if roll_number % 2 == 0:
    df["Electronics"] *= 0.9
else:
    df["Clothing"] *= 0.85
print("Data with discount applied:\n", df)



```

```python
# Plot monthly sales trends for each category using line plots.
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 5))
for category in ["Electronics", "Clothing", "Home & Kitchen", "Sports"]:
    plt.plot(df.index, df[category], label=category)

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trends")
plt.legend()
plt.grid(True)
plt.show()
#Create the following plots:
# Box plots to show the sales distribuƟon for each category.
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[["Electronics", "Clothing", "Home & Kitchen", "Sports"]])
plt.title("Sales Distribution by Category")
plt.ylabel("Sales")
plt.show()
df.to_csv("processed_sales_data.csv", index=True)

```

```python
#WAP to swap two elements in a list in Python Using a Temporary Variable.
def swap_elements(lst, pos1, pos2):
    temp = lst[pos1]
    lst[pos1] = lst[pos2]
    lst[pos2] = temp

my_list = [1, 2, 3, 4, 5]

# Positions to swap (index starts from 0)
pos1, pos2 = 1, 3

print("Before swapping:", my_list)
swap_elements(my_list, pos1, pos2)
print("After swapping:", my_list)
```

```python
#Write a Python program to swap two elements in a set by converƟng it to a list first
def swap_set_elements(s, val1, val2):
    lst = list(s)


    if val1 in lst and val2 in lst:
        idx1, idx2 = lst.index(val1), lst.index(val2)

        # Swap elements in the list
        lst[idx1], lst[idx2] = lst[idx2], lst[idx1]

        # Convert back to set
        return set(lst)
    else:
        print("One or both elements not found in the set!")
        return s
my_set = {10, 20, 30, 40, 50}

# Values to swap
val1, val2 = 20, 40

print("Before swapping:", my_set)
my_set = swap_set_elements(my_set, val1, val2)
print("After swapping:", my_set)



```
